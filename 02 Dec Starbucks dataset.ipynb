```python
## Read the dataset starbucks_final.csv:


#  An ideal diet should contain optimum level of nutrients

# Can you findout the names of the items on menu that contain:
             # upto 450 calories
             # upto 40g protein
             # upto 10g fat
             # upto 40g carbs
             # upto 5g fiber

# Give the names of items on menu that satisfy the above criteria but contain least carbs but max protein.             
```

```python
import pandas as pan
import numpy as np
import os
```

```python
data_directory="D:\\PYTHON JIGSAW"
os.chdir(data_directory)
```

```python
star=pan.read_csv("starbucks_final.csv",encoding="latin")
print(star.head())
```

```python
print(star.shape)
```

```python
# upto 450g calories

star[star["Calories"]<=450].sort_values("Calories",ascending=False)[["Name","Calories"]]

# Answer 84 rows
```

```python
# upto 40g protein

star[star["Protein"]<=40].sort_values("Protein",ascending=False)[["Name","Protein"]]

# Answer 112 rows
```

```python
# upto 10g fat

star[star["Fat"]<=10][["Name","Fat"]].sort_values("Fat",ascending=False)
```

```python
star[star["Fat"]<=10][["Name","Fat"]].sort_values("Fat",ascending=False).shape

# Answer 31 rows
```

```python
# upto 40g carbs

star[star["Carb"]<=40][["Name","Carb"]].sort_values("Carb",ascending=False)
```

```python
star[star["Carb"]<=40][["Name","Carb"]].sort_values("Carb",ascending=False).shape

# Answer 49 rows
```

```python
# upto 5g fiber

star[star["Fiber"]<=5][["Name","Fiber"]].sort_values("Fiber",ascending=False)

# Answer 98 rows
```

```python
# Give the names of items on menu that satisfy the above criteria but contain least carbs but max protein.

star[star["Calories"]<=450 and ["Protein"]<=40 and ["Fat"]<=10 and ["Carb"]<=40 and ["Fiber"]<=5][["Name","Carb","Protein"]] 
```

```python
star.query("Calories<=450 and Protein<=40 and Fat<=10 and Carb<=40 and Fiber<=5")[["Name","Carb"]].sort_values("Carb",ascending=True)
```

```python
star.query("Calories<=450 and Protein<=40 and Fat<=10 and Carb<=40 and Fiber<=5")[["Name","Protein"]].sort_values("Protein",ascending=False)
```
