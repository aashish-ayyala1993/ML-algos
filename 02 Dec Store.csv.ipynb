```python
## Groupby Aggregations

# Average quantity sold in each region
# Different ways are there to calculate
# Way 1

data.groupby("Region",as_index=False).agg({"Quantity":np.mean})
```

```python
# Way 2
data.groupby("Region",as_index=False).agg({"Quantity":"mean"})
```

```python
# Way 3

data.groupby("Region",as_index=False)["Quantity"].mean()
```

```python
#Way 4

data.groupby("Region",as_index=False)["Quantity"].agg("mean")
```

```python
# Calculate both Avg. Sales and Total Sales for each region


data.groupby("Region",as_index=False).agg({"Sales":[np.mean,np.sum]}) 
```

```python
data.groupby("Region",as_index=False).agg({"Sales":[np.mean,np.sum]}).reset_index()
```

```python
# Total Sales and Max profit by Region

data.groupby("Region",as_index=False).agg({"Sales":np.mean,"Profit":np.max})
```

```python
# Total sales and max profit by each Segment

data.groupby("Segment",as_index=False).agg({"Sales":np.sum,"Profit":np.max})
```

```python
# Arrange by profits



data.groupby("Segment",as_index=False).agg({"Sales":np.sum,"Profit":np.max,"Quantity":np.sum}).sort_values("Profit",ascending=False)
```

```python
# How much time will it take to place an order and ship it

data["Order Date"]=pan.to_datetime(data["Order Date"])
data["Ship Date"]=pan.to_datetime(data["Ship Date"])

data["Duration"]=data["Ship Date"]-data["Order Date"]
data["Duration"].head()
```

```python
# How many times the order has been shipped after 5 days from the date when order was placed

data[data["Duration"]>"5 Days"][["Order ID","Duration"]]
```

```python
# How many orders were placed during each day of the week 

data["Weekdays1"]=data["Order Date"].dt.weekday
data.groupby("Weekdays1").size()
```
